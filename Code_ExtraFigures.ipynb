```python
import matplotlib.pyplot as plt
import numpy as np

# Data
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
xgboost_values = [85.93, 85.98, 85.93, 85.94]
lightgbm_values = [85.61, 85.60, 85.61, 85.60]
gbm_values = [84.65, 84.72, 84.65, 84.65]

x = np.arange(len(metrics))  # the label locations

fig, ax = plt.subplots(figsize=(12, 8))

# Plot lines with markers
ax.plot(x, xgboost_values, marker='o', label='XGBoost', color='#1f77b4', linestyle='-')
ax.plot(x, lightgbm_values, marker='s', label='LightGBM', color='#ff7f0e', linestyle='--')
ax.plot(x, gbm_values, marker='^', label='Gradient Boosting (GBM)', color='#2ca02c', linestyle='-.')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Metric Value (%)', fontsize=14)
ax.set_title('Performance Metrics by Model', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(metrics, fontsize=12)
ax.set_yticks(np.arange(84, 90, 1))
ax.set_ylim([84, 90])  # Set y-axis limit for better comparison

# Add grid for better readability
ax.grid(True, linestyle='--', alpha=0.7)

# Adjust legend
ax.legend(loc='upper right', fontsize=12)

# Add percentage values on the markers
for i in range(len(metrics)):
    ax.annotate(f'{xgboost_values[i]:.2f}%', xy=(i, xgboost_values[i]), xytext=(5, 5), textcoords='offset points', fontsize=12, color='#1f77b4')
    ax.annotate(f'{lightgbm_values[i]:.2f}%', xy=(i, lightgbm_values[i]), xytext=(5, -15), textcoords='offset points', fontsize=12, color='#ff7f0e')
    ax.annotate(f'{gbm_values[i]:.2f}%', xy=(i, gbm_values[i]), xytext=(-25, -15), textcoords='offset points', fontsize=12, color='#2ca02c')

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['XGBoost', 'LightGBM', 'GBM']
train_time = [61.08, 55.99, 162.98]
predict_time = [0.13, 0.90, 0.13]
interpretability = ['Medium', 'Medium-High', 'High']
interpretability_scores = [2, 3, 4]  # Mapping interpretability to numeric values

# Training Time Bar Chart
plt.figure(figsize=(12, 8))
plt.bar(models, train_time, color='skyblue', edgecolor='grey')
plt.ylabel('Train Time (s)')
plt.title('Training Time of Models')
plt.ylim(0, 180)
plt.tight_layout()

# Adding value labels for training time
for i, v in enumerate(train_time):
    plt.text(i, v + 2, str(v), ha='center', va='bottom', fontsize=10)

plt.show()

# Prediction Time Bar Chart
plt.figure(figsize=(12, 8))
plt.bar(models, predict_time, color='salmon', edgecolor='grey')
plt.ylabel('Prediction Time (s)')
plt.title('Prediction Time of Models')
plt.ylim(0, 1)
plt.tight_layout()

# Adding value labels for prediction time
for i, v in enumerate(predict_time):
    plt.text(i, v + 0.02, str(v), ha='center', va='bottom', fontsize=10)

plt.show()

# Interpretability Bar Chart
plt.figure(figsize=(12, 8))
interpretability_colors = ['lightgreen', 'lightblue', 'lightcoral']
plt.bar(models, interpretability_scores, color=interpretability_colors, edgecolor='grey')
plt.ylabel('Interpretability (Score)')
plt.title('Interpretability of Models')
plt.ylim(0, 5)
plt.yticks([1, 2, 3, 4, 5], ['Very Low', 'Low', 'Medium', 'Medium-High', 'High'])

# Adding interpretability text annotations
for i, v in enumerate(interpretability_scores):
    plt.text(i, v + 0.1, interpretability[i], ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
